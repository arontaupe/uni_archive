### 2 - Feature Visualization

```python
# Feature visualization by optimization creates images that excite a certain neuron or channel (feature map).

# We start with a random image.
rnd_img = np.random.normal(scale=1,size=(224,224,3))
rnd_img = rnd_img.astype(np.float32)

# Preprocess.
rnd_img = tf.nn.sigmoid(rnd_img)
rnd_img = (rnd_img-0.5)*2.0
rnd_img = tf.convert_to_tensor(rnd_img,dtype=np.float32)

# Plot.
plt.figure(figsize=(7,7))
plt.imshow(tf.cast(255*(rnd_img + 1.0)/2.0, tf.uint8))
plt.axis('off')
plt.show()
```

```python
# To optimize a specific channel we need the feature map as the output of our model.
# We don't care about the rest.

# First we define the channel we are interested in.
feature_map = model.get_layer('mixed3').output[:,:,:,5]

# Then we define a new model that returns this channel as an output.
visualization_model = tf.keras.Model(
                            inputs=model.inputs, 
                            outputs=feature_map
                        )
```

```python
# Hyperparameters.
LEARNING_RATE = 0.03
NUMBER_OF_STEPS = 1000

# Update loop.
for step in range(NUMBER_OF_STEPS):
    
    with tf.GradientTape() as tape:
        
        # The gradient tape will not care about the image, because usually it is not a variable.
        # GradientTape has a method called 'watch', which can be used here.
        ### YOUR CODE HERE ###
        tape.watch(rnd_img)
        ######################
        
        # Feed the image (remember to add batch dimension) to the model and compute the
        # average activation of the feature map. To get the average use 'tf.reduce_mean'.
        ### YOUR CODE HERE ###
        img_batch = tf.expand_dims(rnd_img, axis=0)
        activations = visualization_model(img_batch)
        average_activation = tf.reduce_mean(activations)
        ######################
        
        # Compute the gradients to maximize the average activation in respect to the image.
        # For that understand how the GradientTape method 'gradient' operates.
        ### YOUR CODE HERE ###
        gradients = tape.gradient(average_activation,rnd_img)
        ######################
    
    # Normalize gradients.
    gradients /= tf.math.reduce_std(gradients) + 1e-8
    
    # Apply gradients to the image.
    ### YOUR CODE HERE ###
    rnd_img = rnd_img + gradients * LEARNING_RATE
    ######################
    
    # Make sure values stay between -1 and 1.
    rnd_img = tf.clip_by_value(rnd_img, -1, 1)

    # Plot visualized feature all 50 steps:
    if step%50 == 0:
        plt.figure(figsize=(5,5))
        plt.axis('off')
        img_show = tf.cast(255*(rnd_img + 1.0)/2.0, tf.uint8)
        plt.imshow(img_show)
        plt.title(average_activation.numpy())
        plt.show()
        
        
    # DON'T EXPECT NICE VISUALIZSATIONS. AS LONG AS YOU SEE SOME PATTERNS EMERGING THAT'S FINE!
    # YOU CAN TRY OUT DIFFERENT CHANNELS TO SEE THE DIFFERENCE!
```

As mentioned in the lecture this naive way of optimizing an image does not produce good images.
There are several regularizers to make the images look better. See https://distill.pub/2017/feature-visualization/ for an overview. In most cases you would need to define these regularizers and add them to the loss.

Luckily the authors of the article also provide an extensive TensorFlow library that allows you to play around with many ways of visualization by optimization: https://github.com/tensorflow/lucid

### 3 - Adversarial Attacks 

```python
# Because the InceptionV3 model is actually quite robust against adversarial attacks we take a simpler
# version of it called MobileV2.
model = tf.keras.applications.MobileNetV2(include_top=True, weights='imagenet')
```

```python
# Compute Adversarial Noise
with tf.GradientTape() as tape:
    
    # Make sure that the variable 'img' still has the labrador from before.
    
    ### YOUR CODE HERE ###
    # Make sure the gradient tape records gradients for image.
    tape.watch(img)
    # Add a batch dimension to the image
    img_batch = tf.expand_dims(img, axis=0)
    # Get the predictions from the model and read out the corresponding label.
    probabilities = model(img_batch)
    label = np.argmax(probabilities[0])
    # Define the loss as -log(probability(true_label))
    loss = -tf.math.log(probabilities[:,label])
    # Compute the gradients.
    gradients = tape.gradient(loss,img)
    # Compute the signed gradients. Use 'tf.math.sign'.
    sign_gradients = tf.math.sign(gradients)
    ######################
    
# Plot.
plt.figure(figsize=(7,7))
plt.imshow(sign_gradients)
plt.axis('off')
plt.title('Adversarial Noise')
plt.show()
```

```python
# Add noise to image for different epsilons
epsilons = [0, 0.07, 0.2, 0.3]

for eps in epsilons:
    
    ### YOUR CODE HERE ###
    # Create attacked image.
    attack_image = img_batch + eps*sign_gradients
    # Make sure value stay between -1 and 1. Use function from feature visualization part (clip...).
    attack_image = tf.clip_by_value(attack_image, -1, 1)
    # Get prediction and confidence of model.
    probabilities= model(attack_image)
    label = np.argmax(probabilities[0])
    confidence = probabilities.numpy()[0,label]
    category = lbl2txt[label]
    ######################
    
    # Plot result.
    plt.figure(figsize=(9,9))
    plt.imshow((attack_image[0]+1.0)/2.0)
    plt.title("Espilon:{0}-{1}, Confidence: {2:1.2f}%".format(eps,category,confidence*100))
    plt.axis('off')
    plt.show()
```
