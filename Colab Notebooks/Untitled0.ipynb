```
import numpy as np
a = np.array([3,2,1])
b = np.array([7,2,1])
c = np.array([4,2,2])
p = np.array([6,2,1])

from numpy.linalg import norm

def area(a, b, c) :
    return 0.5 * norm( np.cross( b-a, c-a ) )
```

```
abc = area(a,b,c)
cap = area(c,a,p)
abp = area(a,b,p)
bcp = area(b,c,p)
print(abc)
alpha = cap/abc
beta = abp/abc
gamma = bcp/abc
print(alpha)
print(beta)
print(gamma)
print(alpha+ beta +gamma == 1)
```
