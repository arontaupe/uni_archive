```
import numpy as np
import matplotlib as mpl
import pandas as pd
import statsmodels
from statsmodels import tsa
from statsmodels.tsa import seasonal
import seaborn
#import pmdarima
```

```
df = pd.read_csv("community_size.csv", parse_dates = [0], index_col = 0)
```

```
df.plot()
```

```

result_add = statsmodels.tsa.seasonal.seasonal_decompose(df[' Users'], model='additive', extrapolate_trend='freq')
```

```
 y = df[' Users'].values -  result_add.trend - result_add.seasonal
pd.plotting.autocorrelation_plot(y)
```

```
 z = df[' Users'].values -  result_add.trend 
 pd.plotting.autocorrelation_plot(z)
```

```
df = pd.read_csv("airline_passengers.csv", parse_dates = [0], index_col = 0)
df = df - df.shift(1)- df.shift(12)+ df.shift(13)

df.plot()
```

```
result_mult = statsmodels.tsa.seasonal.seasonal_decompose(df[df.columns[0]], model='multiplicative', extrapolate_trend='freq')
y = df[df.columns[0]].values /  result_mult.trend / result_mult.seasonal
pd.plotting.autocorrelation_plot(y)
```
